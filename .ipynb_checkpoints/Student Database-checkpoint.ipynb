```python
#create the lists with the student information
name = ["Brian Levesque", "Travis Kelce", "Drake Graham", "Sam Howell"]
hometown = ["Waldorf", "Cleveland", "Toronto", "Charlotte"]
fav_food = ["tacos","pizza", "poutine", "bbq"]

acceptable_hometown = ["hometown", "home", "town"]
acceptable_favorite_food = ["favorite food", "food", "favorite"]

#creating a boolean for to repeat the questions
repeat = True

#creating the index variable that's intial value is invalid to catch invalid responses
student = 0

while repeat:
    
    #Provide an option where the user can see a list of all students.
    s_list = input("Would you like to see a list of students? Enter 'y' or 'n'  ")
    if s_list == "y" or s_list == "Y":
        print(name)
    
    #Prompt the user to ask about a particular student by number
    stud = input(f"Which student would you like to know more about (please enter their number 1-{len(name)}) or name: ")
    
    #checking if input was a student name
    if stud in name:
        student = name.index(stud) + 1
    else:
        if stud.isnumeric():
            student = int(stud)
    
    #Validate user number
    if student >= 1 and student <= len(name):
        
        #Print the studentâ€™s name.
        print(f'Student {student} is {name[student-1]}')
    
        #Ask the user which category to display: Hometown or Favorite food.
        cate_loop = True
        
        while cate_loop:
            
            #Ask the user which category to display: Hometown or Favorite food.
            category = input("Which category do you want to see Hometown or Favorite food? ")
            
            #make the input is in constant case for checking purposes
            category = category.lower()
            
            #display the relevant information.
            if category in acceptable_hometown:
                print(hometown[student-1])
                cate_loop = False
        
            elif category in acceptable_favorite_food:
                print(fav_food[student-1])
                cate_loop = False
            
            else:
                print("Please enter valid category.")
        
        #Ask the user if they would like to learn about another student.

        again = input('Would you like to learn about another student? Enter "y" or "n" ')
        if again == "y" or again == "Y":
            repeat = True
            print("Ok.")
        else:
            repeat = False
            print("Thanks!")
    else:
        print(f'Please enter a valid student name or number (1-{len(name)}.) ')
    

```
